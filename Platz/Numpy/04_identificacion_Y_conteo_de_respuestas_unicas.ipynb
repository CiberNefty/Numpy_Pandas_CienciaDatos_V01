**Unique**

```python
import numpy as np
```

```python
survey_responses = np.array(['bueno',"excelente",'malo',
                             "bueno","malo","excelente",
                             "bueno","bueno","malo","excelente"])

print(np.unique(survey_responses))
```

```python
# Cuantas veces aparece un elemento:
unique_elements, indiceElemento, counts = np.unique(survey_responses, return_counts=True, return_index=True)
print(f"Elementos unicos: {unique_elements}")
print(f"Indice de cada elemento: {indiceElemento}")
print(f"Cuantos elementos tiene cada uno: {counts}")
```

**Copia / Visatas**

```python
# Vistas
array_x = np.arange(10)
view_y = array_x[1:3]

print(array_x)
print(view_y)

# Cambiar elementos desde una posicion
array_x[1:3] = [10,11]
print(array_x)
print(view_y)
```

```python
# Copia
array_x = np.arange(10)
copy_x = array_x[1:3].copy()
#copy_x = array_x[[1,2]]
print(array_x)
print(copy_x)
array_x[1:3] = [10,11]
print(array_x)
print(copy_x)
```

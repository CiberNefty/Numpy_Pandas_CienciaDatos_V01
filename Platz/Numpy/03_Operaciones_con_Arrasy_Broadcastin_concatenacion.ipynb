**Broadcasting**

```
Es cuando NumPy expande arrays automáticamente para poder operar aunque no tengan la misma forma.
```



````python
import numpy as np
````

````python
prices = np.array([100,200,300])
discount = np.array([0.9])
discount_price = prices*discount
print(discount_price)
````

````python
prices = np.random.randint(100,500, size=(3,3))
discount = np.array([10,20,30])
discount_price = prices+discount
print(prices)
print(discount_price)
````

**Operaciones logicas**

````python
array = np.array([1,2,3,4,5])
print(np.all(array > 91))
print(np.any(array > 2))
````

**Concatenacion**

````python
array_a = np.array([1,2,3])
array_b = np.array([4,5,6])
concatenated_a = np.concatenate((array_a, array_a))
concatenated_ab = np.concatenate((array_a, array_b))
print(concatenated_a)
print(concatenated_ab)
````

**Staking**:

Una forma de apilar arrays

````python
satecked_v = np.vstack((array_a,array_b))
print(f"vertical:\n{satecked_v}")
satecked_h = np.hstack((array_a,array_b))
print(f"Horizontal: {satecked_h}")
````

**Split**

````python
array_c = np.arange(1,10)
split_array = np.split(array_c,3)
print(split_array)
````

**Ejercicios:**

````python
# 1. Broadcasting

a = np.array([[1],[2],[3]])
b = np.array([10,20,30])
print(a+b)
````

````python
# 2. Concatenacion
# Une arrays en un eje específico.
a = np.array([[1,2],[3,4]])
b = np.array([[5,6]])
concatenated = np.concatenate((a, b), axis=0)
print(concatenated)
````

````python
# 3. Staking (vstack, hstack)
# Es como concatenar pero mas visual

a = np.array([[1],[2],[3]])
b = np.array([[10],[20],[30]])
print(np.hstack((a,b)))
print("\n")
print(np.vstack((a,b)))
````

````python
# 4. Split
# Divide arrays en partes iguales.

x = np.arange(16).reshape(4,4)
print(f"Filas:\n{np.split(x, 2)}")
print(f"Columnas:\n{np.split(x, 2, axis=0)}")

# All y Any
print(np.all(x > 20))
print(np.any(x > 12))
````

````python
# 5. Mesclando todo

a = np.arange(1,10).reshape(3,3)
b = np.array([1,2,3])

c = a * b              # broadcasting
d = np.hstack((a,c))  # stacking
e = np.split(d, 3)    # split

print(c)
print(d)
print(e)
````

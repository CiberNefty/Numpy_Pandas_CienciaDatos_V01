```python
import numpy as np
```

```python
# Paso 1
months = np.array(['Enero','Febrero','Marzo','Abril','Mayo','Junio','Julio',
                   'Agosto','Septiembre','Octubre','Noviembre','Diciembre'])
ventas_pruducto_A = np.array([150,200,250,300,220,210,180,190,230,240,280,300])
ventas_pruducto_B = np.array([180,210,230,250,270,260,240,250,270,290,310,330])
ventas_pruducto_C = np.array([200,220,240,260,280,300,320,340,360,380,400,420])
```

```python
# Paso 2: Resuelto

# Suma de ventas a,b,c totales
ventas_totales = np.sum([ventas_pruducto_A,ventas_pruducto_B,ventas_pruducto_C])
print(f"Ventas totales de los productos A,B,C: ${ventas_totales}")


# Media de las ventas del ´producto A:
media_prodcutoA = np.mean(ventas_pruducto_A)
print(f"\nMedia de las ventas del producto A: {media_prodcutoA}")
print(f"Suma de las ventas del producto A: {np.sum(ventas_pruducto_A)}")
# Media de las ventas del ´producto B:
media_prodcutoB = np.mean(ventas_pruducto_B)
print(f"\nMedia de las ventas del producto B: {media_prodcutoB}")
print(f"Suma de las ventas del producto B: {np.sum(ventas_pruducto_B)}")
# Media de las ventas del ´producto C:
media_prodcutoC = np.mean(ventas_pruducto_C)
print(f"\nMedia de las ventas del producto C: {media_prodcutoC}")
print(f"Suma de las ventas del producto C: {np.sum(ventas_pruducto_C)}")
```

```python
# Paso 3: Resuelto
"""
Reshape y Transposicion:

Crea una matriz 2D con las ventas de los tres productos y transforma su forma(resahpe)
a un array tridimensional con dimensiones (3,4,4)
Trasnpone la matriz de ventas para que las filas se conviertan en columnas.
Invierte arrays: Invierte las ventas de cada producto en orden de meses.
Aplanar la matriz: Convierte la matriz a un array unidimensional.
"""
matrix_2d = np.array([ventas_pruducto_A,ventas_pruducto_B,ventas_pruducto_C])
matrix_reshape = matrix_2d.reshape(3,4,3)
matrix_transpose = matrix_2d.transpose()
matrix_inversa_meses = matrix_2d[:, ::-1]
matrix_aplanada = matrix_2d.flatten()
print(f"Matriz todas las ventas 2D:\n {matrix_2d}\n")
print(f"Forma Reshape 3x4x4:\n {matrix_reshape}\n")
print(f"Forma transponer de filas a columnas:\n {matrix_transpose}\n")
print(f"Forma matrix inversa por meses:\n {matrix_inversa_meses}\n")
print(f"Forma de matrix aplanada:\n {matrix_aplanada}")



```

```python
# Paso 4:
"""
Utiliza np.unique para encontrar los elementos unicos en los datos de ventas y
cuenta cuantas veces aparece cada uno."""
print(f"Matriz ventas totales=\n {matrix_2d}")
elementosUnicos, conteoElementosUnicos = np.unique(matrix_aplanada, return_counts=True)
print(f"\nElementos Unicos en nuestra matriz de ventas:\n {elementosUnicos}")
print(f"\nConteo de cada elemento unico:\n {conteoElementosUnicos}\n")


# Solucion con FOR:

for elemento in elementosUnicos:
  conteo = (matrix_aplanada == elemento).sum()
  print(f"{elemento} aparece: {conteo}")
```

```python
# Paso 5:
"""Indexacion y Slicing
Ventas de primer trimestre: Extrae las ventas de los primeros meses del año.
Indexacion booleana: Selecciona los meses donde el total de ventas de
los tres productos supere los 800. Seleccion avanzada: usa una lista de indices
 para seleccionar las ventas de los meses pares (o selecciona los meses a tu
 eleccion) y muestras esas ventas en una nueva matriz."""

primer_trimestre = matrix_2d[:,:3]
print(f"Ventas del año por cada producto:\n{matrix_2d}")
print(f"\nVentas Primer Trimestre:\n{primer_trimestre}")
ventas_trimestre1 = np.sum(matrix_2d[:,:3])

print(f"\nVenta total primero trimestre: ${ventas_trimestre1}")
inicio =0
fin =1
print(f"\nMeses donde el total de ventas supero los $800:")
for mes in range(12):
  #valor= np.sum(matrix_2d[:,mes]) # Respuesta igual
  valor= np.sum(matrix_2d[:,inicio:fin])
  inicio+=1
  fin+=1
  index_bool = valor >= 800
  if index_bool:
    print(f"Mes # {mes+1}: ${valor}")

lista_indices = []

for mes in range(12):
  if mes % 2 == 0:
    lista_indices.append(mes)

print(f"\nMeses pares: {lista_indices}")
new_matrix = matrix_2d[:, lista_indices]
print(f"\nNueva matriz con meses pares seleccionados:\n{new_matrix}")
```

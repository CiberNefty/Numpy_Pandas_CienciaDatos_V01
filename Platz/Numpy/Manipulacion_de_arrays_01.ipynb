**Manipulacion de Arrays**

```python
import numpy as np
```

```python
array = np.array([[1,2,3],[4,5,6]])
print(array.ndim) # Dimension del array
print(array.shape) # un vector de 2 x 3
print(array.dtype) # Tipo de dato int64
```

```python
z = np.array(3, dtype= np.uint8)
print(z)
```

```python
double_array = np.array([1,2,3], dtype='d')
print(double_array)
```

```python
z = z.astype(np.float64)
print(z)
```

```python
# SUMA
array = np.array([[1,2,3],[4,5,6]])
add = np.sum(array)
print(add)
```

```python
# MEDIA
mean = np.mean(array)
print(mean)
```

```python
# DESVIACION ESTANDAR
std = np.std(array)
print(std)
```

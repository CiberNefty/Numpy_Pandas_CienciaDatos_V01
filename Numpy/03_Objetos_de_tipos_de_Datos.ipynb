```python
import numpy as np
```

```python
# Un tipo de dato sencillo que contiene un entero big-endian de 32 bits
dt = np.dtype('>i4')
print(dt.byteorder)
print(dt.itemsize)
print(dt.name)
print(dt.type is np.int32)
```

```python
# Un tipo de datos estructurado que contiene una cadena de 16 caracteres
# (en el campo 'nombre') y un subarray de dos números de coma flotante de 64 bits (en el campo 'grades'):
dt  = np.dtype([('name', np.str_, 16), ('grades', np.float64, (2,))])
print(dt['name'])
print(dt['grades'])
```

```python
# Los elementos de un array de este tipo de datos se envuelven en un array tipo escalar que también tiene dos campos:
x = np.array([('Sarah',(8.0, 7.0)), ('John',(6.0, 7.0))], dtype=dt)
print(x[1])
print(x[1]['grades'])
print(type(x[1]))
print(type(x[1]['grades']))
```

```python
# =========================================================================================
## Tipos de datos con .dtype

# Cadenas de un solo carácter
# Cada tipo de dato incorporado tiene un código de carácter (los códigos de tipo numéricos actualizados), que lo identifica de forma única.

dt = np.dtype('b')  # byte, native byte order
dt = np.dtype('>H') # big-endian unsigned short
dt = np.dtype('<f') # little-endian single-precision float
dt = np.dtype('d')  # double-precision floating-point number

# Cadenas de tipo protocolo de array
dt = np.dtype('i4')   # 32-bit signed integer
dt = np.dtype('f8')   # 64-bit floating-point number
dt = np.dtype('c16')  # 128-bit complex floating-point number
dt = np.dtype('S25')  # 25-length zero-terminated bytes
dt = np.dtype('U25')  # 25-character string
```

```python
# Cadena con campos separados por comas

# campo nombrado que contiene un entero de 32 bitsf0
# campo nombrado que contiene un subarray de 2 x 3 de números de coma flotante de 64 bitsf1
# campo nombrado que contiene un número de coma flotante de 32 bitsf2

dt = np.dtype("i4, (2,3)f8, f4")
# campo nombrado que contiene una cadena de 3 caracteresf0
# campo nombrado que contiene un subarray de forma (3,) que contienen enteros sin signo de 64 bitsf1
# campo nombrado que contiene un subarray de 3 x 4 que contienen cadenas de 10 caracteresf2

dt = np.dtype("S3, 3u8, (3,4)S10")

# Cuerdas tipo: Cualquier nombre de cadena de un tipo NumPy dtype, por ejemplo:
dt = np.dtype('uint32')   # 32-bit unsigned integer
dt = np.dtype('float64')  # 64-bit floating-point number
```

{'names': ..., 'formats': ..., 'offsets': ..., 'titles': ..., 'itemsize': ...}
Este estilo tiene dos llaves obligatorias y tres opcionales. Se requieren las claves de nombres y formatos. Sus respectivos valores son listas de longitud igual con los nombres y formatos de campos. Los nombres de los campos deben ser cadenas y los formatos de campo pueden ser cualquiera objeto aceptado por el constructor.

Cuando se proporcionan los desplazamientos y títulos opcionales de las teclas, sus valores deben ser listas de la misma longitud que las listas de nombres y formatos. El valor de los desplazamientos es una lista de desplazamientos de bytes (limitado a ) para cada campo, mientras que el valor de los títulos es un Lista de títulos para cada campo ( puede usarse si no hay título deseado para ese campo). Los títulos pueden ser cualquier objeto, pero cuando un objeto añade otra entrada a la Diccionario de campos con el título y referenciando el mismo Tupla de campo que contendrá el título como una tupla adicional miembro.None

La clave de tamaño del ítem permite que el tamaño total del dtype sea y debe ser un entero lo suficientemente grande como para que todos los campos están dentro del dtype. Si el dtype que se está construyendo está alineado, El tamaño del objeto también debe ser divisible por la alineación de la estructura. El tamaño total del objeto dtype está limitado a .

```python
# Tipo de dato con campos , , , , cada uno siendo un entero sin signo de 8 bits:rgba
import numpy as np
```

```python
dt = np.dtype({'names': ['r','g','b','a'],
               'formats': [np.uint8, np.uint8, np.uint8, np.uint8]})
```

```python
# Tipo de dato con campos y (con los títulos indicados), ambos son enteros sin signo de 8 bits, el primero en posición de byte 0 desde la salida del pelotón y el segundo en la posición 2:rb
dt = np.dtype({'names': ['r','b'], 'formats': ['u1', 'u1'],
               'offsets': [0, 2],
               'titles': ['Red pixel', 'Blue pixel']})

print(dt)
```

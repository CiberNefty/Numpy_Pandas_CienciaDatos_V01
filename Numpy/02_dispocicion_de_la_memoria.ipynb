**Disposición de la memoria**
*Los siguientes atributos contienen información sobre la disposición de la memoria Del arreglo:*

```python
# ndarray.shape
import numpy as np
x = np.ndarray([1,2,3,4,5])
print(x.shape)
y = np.zeros((2,3,4))
print(y.shape)
y.shape = (3,8)
print(y)
```

```python
# ndarray.strides
import numpy  as np
y = np.reshape(np.arange(2*3*4, dtype=np.int32),(2,3,4))
print(y)
print(y.strides)
print(y[1,1,1])
offset = sum(y.strides * np.array((1,1,1)))
offset // y.itemsize
```

```python
x = np.reshape(np.arange(5*6*7*8, dtype=np.int32), (5, 6, 7, 8))
x = x.transpose(2,3,1,0)
print((x.strides))

i = np.array([3,5,2,2], dtype = np.int32)
offset = sum(i * x.strides)
print(x[3,5,2,2])

print(offset // x.itemsize)
```

```python
# ndarray.ndim
# 	Número de dimensiones de la matriz.
import numpy as np
x = np.array([1,2,3])
print(x.ndim)

y = np.zeros((2,3,4))
print(y.ndim)

# ndarray.data
#   Objeto búfer de Python apuntando al inicio de los datos del array.
print(x.data)
```

```python
# ndarray.size
#   Número de elementos en el arreglo.
import numpy as np
x = np.zeros((3,5,2), dtype=np.complex128)
x.size
```

```python
# ndarray.itemsize
#   Longitud de un elemento del array en bytes.
import numpy as np
x = np.array([1,2,3],dtype= np.float64)
print(x.itemsize)

x = np.array([1,2,3], dtype = np.complex128)
x.itemsize
```

```python
# ndarray.nbytes
#   Total de bytes consumidos por los elementos del array.
import numpy as np
x = np.zeros((3,5,2), dtype=np.complex128)
print(x.nbytes)

print(np.prod(x.shape) * x.itemsize)
```

```python
# ndarray.base
#   Objeto base si la memoria proviene de otro objeto.
import numpy  as np
x = np.array([1,2,3,4])
print(x.base is None)

y = x[2:]
y.base is x
```
